```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
occurrences = pd.read_csv('./database/ocorrencia.csv', delimiter=';')
occurrences
```

```python
airplanes = pd.read_csv('./database/aeronave.csv', delimiter=';')
airplanes
```

```python
occurrences_complete = pd.merge(occurrences, airplanes, on=['codigo_ocorrencia2'])
occurrences_complete
```

```python
occurrences_complete
```

```python
occurrences_complete.info()
```

```python
sns.barplot(x='ocorrencia_uf', y='aeronave_nivel_dano', hue='aeronave_nivel_dano', data=occurrences_complete)
```
