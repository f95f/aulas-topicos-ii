```python
grupo_classe = df.groupby('pclass')
grupo_classe.first()
```

```python
idade_mediana_por_classe = grupo_classe['age'].median()
idade_mediana_por_classe
```

```python
def aplicar_idade_se_vazio(x):
    if pd.isna(x['age']):
        return idade_mediana_por_classe[x['pclass']]
    else:
        return x['age']

df_idade_completa['age'] = df.apply(aplicar_idade_se_vazio, axis=1)
```
