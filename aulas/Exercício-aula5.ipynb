```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
#histograma da coluna tips
sns.histplot(tips['tip'], kde=True)
```

```python
#boxplot da coluna total_bill
sns.boxplot(x='total_bill', y='day', data=tips, hue='day')
```

```python
#boxplot usando as variÃ¡veis tips e time
sns.boxplot(x='time', y='tip', data=tips, hue='time')
```

```python
#violinplot usando as variÃ¡veis tips e smoker
sns.violinplot(x='smoker', y='tip', data=tips, hue='smoker')
```
